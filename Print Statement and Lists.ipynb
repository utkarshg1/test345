# Variable substitution
1. Comma Method
2. f string (f - format)
3. .format()



```python
# Comma method
name1 = "Sarthak"
mob1 = 9876543210
bal1 = 0.35

print("Hi", name1, "your mobile number :", mob1, "is due for recharge")
print("Current balance is", bal1, "INR")
```

```python
name2 = "Raman"
mob2 = 8721321421
bal2 = 0.21312421

print("Hi", name2, "your mobile number :", mob2, "is due for recharge")
print("Current balance is", round(bal2, 4), "INR")
```

# fstring method
1. starts with letter f
2. variable should be inside placeholder {}

```python
name3 = "Priya"
mob3 = 213124214
bal3 = 0.123214123

print(f"Hi {name3}, Your mobile number : {mob3} is due for recharge")
print(f"Current balance is {bal3:.3f} INR")
```

# .format method

```python
name4 = "Rahul"
mob4 = 5324213412
bal4 = 0.512321321

print("Hi {}, Your mobile number {} is due for recharge".format(name4, mob4))
print("Current balance is {:.2f} INR".format(bal4))
```

```python
print("Hi {name}, Your mobile number {mob} is due for reacharge".format(mob = mob4, name = name4))
print("Current balance is {bal:.2f} INR".format(bal = bal4))
```

# Calculate simple interest and amount for given p, n and r

```python
p = float(input("Enter Principal in INR : "))
n = int(input("Enter number of years : "))
r = float(input("Enter rate of interest in %p.a. : "))

i = (p * n * r) / 100
a = p + i

print(f"Simple Interest : {i:.2f} INR")
print(f"Amount : {a:.2f} INR")
```
